```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')

plt.show()
```

```python
# Look at correlation between the columns

# Types of correlations
# pearson, kendall, spearman
```

```python

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix Viz for numeric features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')

plt.show()
```

```python
# Unstack the correlation into table for easy view in pairs

corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
# Sort the correlation table into sorted pairs

corr_pairs_sorted = corr_pairs.sort_values()

corr_pairs_sorted
```
